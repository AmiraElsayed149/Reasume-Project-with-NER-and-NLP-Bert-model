```python
def get_skills(text):
    doc = nlp(text)
    #print(doc)
    myset = []
    subset = []
    for ent in doc.ents:
        #print("1")
        if ent.label_ == "SKILL":
            subset.append(ent.text)
    myset.append(subset)
    print(subset)
    return subset


def unique_skills(x):
    return list(set(x))
```

```python
'''
The optimizations include:

Moving the initialization of the stop words and lemmatizer objects outside the loop, since they don't change from iteration to iteration.
Using a for-each loop to iterate over the reviews in df["Resume_str"] instead of using range and iloc.
Removing the redundant assignment of df["Resume_str"].iloc[i] to review in each iteration.
Using set membership (if word not in stop_words) instead of calling set(stopwords.words("english"))._contains_(word) for each word.
Removing unnecessary parentheses and whitespace.
'''
clean = []
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

for review in df["Resume"]:
    # Remove special characters, URLs, and Twitter handles
    review = re.sub('(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?"', " ", review)
    # Convert to lowercase and split into words
    review = review.lower().split()
    # Lemmatize words and remove stop words
    review = [lemmatizer.lemmatize(word) for word in review if word not in stop_words]
    # Join the words back into a single string
    review = " ".join(review)
    clean.append(review)
```

```python
df["Clean_Resume"] = clean
df["skills"] = df["Clean_Resume"].str.lower().apply(get_skills)
print(df["skills"])
df["skills"] = df["skills"].apply(unique_skills)
df
```

```python
%%time
import pandas as pd
#from request import PandaRequest
## extract sentences with skills
## optomized by Nady
import re
df2 = pd.DataFrame(columns=["sentence_id", "word", "word_mdf"])
for row, sentence in enumerate(sentence_listoflists):
  sentence = ' '.join(sentence)
  sentence_cp = sentence
  
  for skill in data_skills:
      if (len(skill.split()) == 1):
        
        for ser in  re.finditer(rf'\b{skill}\b', sentence):
          if ser != None:
            sentence = re.sub(rf"\b{skill}\b", "B-SKILL", sentence)
          else:
            pass
      elif (len(skill.split()) == 2):    
        for ser in re.finditer(rf'\b{skill}\b', sentence):
          if ser != None:
            sentence = re.sub(rf"\b{skill}\b", "B-SKILL I-SKILL", sentence)
          else:
            pass
      elif (len(skill.split()) == 3):    
        for ser in re.finditer(rf'\b{skill}\b', sentence):
          if ser != None:
            sentence = re.sub(rf"\b{skill}\b", "B-SKILL I-SKILL O-SKILL", sentence)
          else:
            pass
  try:
    token_sentence_cp = sentence_cp.split()
    token_sentence = sentence.split()
    sentence_id = [f'sentence_{row}']*len(token_sentence)
    token_df = pd.DataFrame({"sentence_id":sentence_id,"word":token_sentence_cp, "word_mdf": token_sentence})
    df2 = pd.concat([df2, token_df])
  except Exception as e:
    print(f"sentence_#: {row}", len(sentence.split()), len(sentence_cp.split()), sep=",")
    print(sentence, sentence_cp, sep="\n")
```
